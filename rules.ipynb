```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df=pd.read_csv(r"C:\Users\10725997\Downloads\alerts-data.csv")
```

```python
df=df[['ciImpacted','kpi','lastModifiedTime']]
```

```python
df.head(10)
```

```python
df=pd.DataFrame(df)
df
```

```python
df['Date']=pd.to_datetime(df['lastModifiedTime'],format='mixed').dt.date
df.drop('lastModifiedTime',axis=1,inplace=True)
```

```python
df
```

```python
df=df.set_index('Date')
```

```python
df
```

```python
df=pd.get_dummies(df)
```

```python
df
```

```python
grouped_data=df.groupby("Date").sum()
```

```python
print(grouped_data)
```

```python
frequent_itemsets=apriori(grouped_data, min_support=0.1, use_colnames=True)
```

```python
def func1(x):
    mask = x>1
    x[mask]=1
    return x

grouped_data=grouped_data.transform(func1)
grouped_data
```

```python
frequent_itemsets=apriori(grouped_data,min_support=0.1,use_colnames=True)
frequent_itemsets
```
